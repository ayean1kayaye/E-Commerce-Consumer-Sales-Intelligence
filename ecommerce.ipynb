```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load data in dataFrame**

```python
customers=pd.read_csv(r"/content/customers (2).csv")
customers.head(7)
```

```python
orders=pd.read_csv(r"/content/orders (1).csv")
orders.head(7)
```

**Merge the two table**

```python
df = orders.merge(customers, on="CustomerID")
df
```

**Total revenue monthly **

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
monthly = df.groupby(df['OrderDate'].dt.to_period('M'))['Quantity'].sum().reset_index()
monthly.head(5)
```

**Find the Customer life value by quantity**

```python
cust_life_value = df.groupby('Name')['Quantity'].sum().reset_index().sort_values(by='Quantity', ascending=False)
cust_life_value.head(5)
```

**Delay Impact**

```python
delay = df.groupby('Status')['Quantity'].sum().reset_index()
delay.head(4)
```

**visualization of total revenue**

```python
df['Revenue'] = df['Quantity'] * df['UnitPrice']
total_revenue = (df.groupby('PaymentMode')['Revenue'].sum().reset_index().sort_values(by='Revenue', ascending=False))
total_revenue.head(5)
```

```python
plt.figure(figsize=(4,6))
plt.pie(total_revenue['Revenue'], labels=total_revenue['PaymentMode'], autopct='%1.1f%%', startangle=90)
plt.legend(total_revenue['PaymentMode'], title="Payment Mode",loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("Revenue Share by Payment Mode", color='blue')
plt.show()
```

**Monthaly Revenues **

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Calculate Monthly Revenue
df['month_revenue'] = df['Quantity'] * df['UnitPrice']
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
monthly = df.groupby(df['OrderDate'].dt.to_period('M'))['month_revenue'].sum().reset_index()
monthly['OrderDate'] = monthly['OrderDate'].dt.to_timestamp()
monthly['Month'] = monthly['OrderDate'].dt.strftime('%B')


def format_m(value, _):
    return f'{value/1_000_000:.1f}M'

plt.figure(figsize=(10,6))
bars = plt.bar(monthly['Month'], monthly['month_revenue'], color=plt.cm.tab20.colors)


for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f"{height/1_000_000:.2f}M",
             ha='center', va='bottom', fontsize=10, fontweight='bold')


plt.gca().yaxis.set_major_formatter(FuncFormatter(format_m))
plt.title("Monthly Revenue Trend", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Monthly Revenue (in Millions)", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

**Total Revenue by Region**

```python
import pandas as pd
import matplotlib.pyplot as plt
df['Revenue'] = df['Quantity'] * df['UnitPrice']
region_revenue = df.groupby('Region')['Revenue'].sum().reset_index()

plt.figure(figsize=(4,7))
plt.pie(region_revenue['Revenue'], labels=region_revenue['Region'], autopct='%1.1f%%', startangle=140)
plt.legend(region_revenue['Region'], title="Region",loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("Revenue Contribution by Region", fontsize=14)
plt.show()
```

**Total sales - quantity**

```python
import pandas as pd
import matplotlib.pyplot as plt


product_quantity = df.groupby('ProductName')['Quantity'].sum().reset_index()
product_quantity = product_quantity.sort_values(by='Quantity', ascending=False)

plt.figure(figsize=(10,6))
plt.barh(product_quantity['ProductName'], product_quantity['Quantity'], color=plt.cm.Set3.colors)
plt.title("Total Quantity Sold by ProductName", fontsize=14)
plt.xlabel("Total Quantity", fontsize=12)
plt.ylabel("ProductName", fontsize=12)
plt.gca().invert_yaxis()  # highest quantity on top
plt.show()
```

**Total Quantity by Status**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
bars = plt.bar(delay['Status'], delay['Quantity'], color=plt.cm.Paired.colors)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f"{height}", ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.title("Total Quantity by Order Status", fontsize=14)
plt.xlabel("Order Status", fontsize=12)
plt.ylabel("Total Quantity", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
